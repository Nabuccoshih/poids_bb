```python
import csv

with open('poids_bb.csv', 'a', newline='', encoding='utf-8') as data_csv:
    writer=csv.writer(data_csv)
    writer.writerow(['16/08/2023','1130','1015','987','940','970','1110','997'])
    
# Lilas - Rosa - Upsie - Doré - Bleu ciel - Bleu foncé - Ultime
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
 
figure(figsize=(15,15))

data = pd.read_csv('poids_bb.csv')

plt.plot(data.Ultime, marker='o', label='Ultime')
plt.plot(data.Bleu_fonce, marker='s', label='Bleu foncé')
plt.plot(data.Bleu_ciel, marker='p', label='Bleu ciel')
plt.plot(data.Dore, marker='*', label='Dorée')
plt.plot(data.Upsie, marker='X', label='Upsie')
plt.plot(data.Lilas, marker='D', label='Lilas')
plt.plot(data.Date, data.Rosa, marker='d', label='Rose')

plt.gcf().autofmt_xdate()
plt.title('Poids des bébés')
plt.xlabel('Date')
plt.ylabel('Masse en grammes')
plt.legend(loc=2, fontsize=20)
plt.grid()

plt.show()
```
